# Excercise 2

```python
import numpy as np
```

```python
student_data = np.array([[78,85,90],
                         [45,60,55],
                         [88,92,95],
                         [30,40,35],
                         [65,70,75]])
print(student_data)
```

```python
print("Shape of student_data : ",student_data.shape)
print("Total Number of elements : ",student_data.size)
print("Data types  of elements : ",student_data.dtype)
```

```python
print("marks of 2nd student: ",student_data[1])
print("\n")
print("marks of 1st student in 3rd subject: ",student_data[0][2])
print("marks of 2nd student in 3rd subject: ",student_data[1][2])
print("marks of 3rd student in 3rd subject: ",student_data[2][2])
print("marks of 4th student in 3rd subject: ",student_data[3][2])
print("marks of 5th student in 3rd subject: ",student_data[4][2])
print("\n")
print("marks of 1st student in 2nd subject: ",student_data[0][1])
```

```python
#MEANS AND AVERAGES
subject_mean = np.mean(student_data, axis=0)
print("Mean of all subjects : ",subject_mean)

avg_marks = np.average(student_data, axis=0)
print("Average of all subjects : ",avg_marks)
```

```python
#filtering Top performers 
result = student_data[student_data >= 80]
print("marks scored above 80 : ",result)

```

```python
student_data+=5
print(student_data)
```

```python
status = np.where(student_data>=50,1,0)
print(status)
```

```python
student_data = np.where(student_data<50,0,student_data)
print(student_data)
```

```python
avg = np.average(student_data,axis=1)
grades = np.where(avg >= 85, 10,
         np.where(avg >= 70, 8,
         np.where(avg >= 50, 6, 0)))
print(grades)
```

```python
backup = student_data.copy()

student_data[0, 0] = 0

print("Original Marks:\n", student_data)
print("Backup Marks:\n", backup)
```
